```python
import tensorflow.keras as keras
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import cv2
import pickle
```

```python
new_model = tf.keras.models.load_model('./animal_detector')
print(new_model.summary())
```

```python
animals = ["horse", "dog", "elephant", "butterfly", "rooster", "cat", "cow", "tarantula", "squirrel", "sheep"]

X = pickle.load(open("pickle/X_test.pickle", "rb"))
y = pickle.load(open("pickle/y_test.pickle", "rb"))

X_train = np.array(X).reshape(-1, 50, 50, 3)
X_train = tf.keras.utils.normalize(X_train, axis=1)

print(X_train.shape)
```

```python
predictions = new_model.predict(X_train)
```

```python
result = []
t = 0
f = 0

for i in range(len(predictions)):
    if np.argmax(predictions[i]) == y[i]:
        t = t + 1
    else:
        f = f + 1
    result.append([np.argmax(predictions[i]), y[i]])
    
    
x = ['Correct', 'Wrong']
popularity = [t, f]
x_pos = [i for i, _ in enumerate(x)]

fig, ax = plt.subplots()
rects1 = ax.bar(x_pos, popularity, color='b')
plt.ylabel("The answers")
plt.title("\n\n" + "Neural Network with for detecting animals\n")
plt.xticks(x_pos, x)
# Turn on the grid
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
# Customize the minor grid
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')


def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2., 1.05 * height,
                '%f' % float(height),
                ha='center', va='bottom')


autolabel(rects1)
plt.show()
```
